```python
import numpy as np
import pandas as pd
import os
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
# performance evaluation
def macro_f1(target,fit):

    """
    计算得分
    :param target_df: [sn,fault_time,label]
    :param submit_df: [sn,fault_time,label]
    :return:
    """

    weights =  [3  /  7,  2  /  7,  1  /  7,  1  /  7]
    
    macro_F1 =  0.
    for i in  range(len(weights)):
        TP =  sum((target == i) & (fit == i))
        FP =  sum((target != i) & (fit == i))
        FN =  sum((target == i) & (fit != i))
        precision = TP /  (TP + FP)  if  (TP + FP)  >  0  else  0
        recall = TP /  (TP + FN)  if  (TP + FN)  >  0  else  0
        F1 =  2  * precision * recall /  (precision + recall)  if  (precision + recall)  >  0  else  0
        macro_F1 += weights[i]  * F1
        
        # print('Task %d:\n Prcesion %.2f, Recall %.2f, F1 %.2f' % (i+1, precision, recall, F1))
        
    return macro_F1
```

```python
train_data = TabularDataset('C:\\Users\\whchi\\Downloads\\Log-diagnosis-main\\Log-diagnosis-main\\pre_contest\\dataset\\train.csv')
```

```python
from autogluon.core.metrics import make_scorer
ag_macro_f1_scorer = make_scorer(name='macro_f1',
                                 score_func=macro_f1,
                                 optimum=1,
                                 greater_is_better=True)
```

```python
predictor = TabularPredictor(label='label',eval_metric=ag_macro_f1_scorer).fit(train_data=train_data, presets='best_quality')
```

```python
test_data = TabularDataset('C:\\Users\\whchi\\Downloads\\Log-diagnosis-main\\Log-diagnosis-main\\pre_contest\\dataset\\test.csv')
predictor.leaderboard(test_data, silent=True)
```
